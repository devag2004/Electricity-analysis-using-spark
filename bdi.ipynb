```python
# Initialize Spark Session
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, mean, min, max, count, desc
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.clustering import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import builtins
from google.colab import files

def create_spark_session():
    """Create and configure a Spark session."""
    spark = SparkSession.builder \
        .appName("ElectricityAnalysis") \
        .config("spark.ui.port", "4050") \
        .config("spark.executor.memory", "2g") \
        .config("spark.driver.memory", "2g") \
        .master("local[*]") \
        .getOrCreate()

    spark.sparkContext.setLogLevel("ERROR")
    return spark

# Data Processor Class
class ElectricityDataProcessor:
    def __init__(self, spark, data_path=None, df=None):
        """Initialize with either a path to data or a DataFrame."""
        self.spark = spark
        self.data_path = data_path
        if df is not None:
            self.spark_df = df
        elif data_path is not None:
            self.load_data()
        else:
            self.spark_df = None

    def load_data(self):
        """Load data from CSV into Spark DataFrame."""
        print(f"Loading data from: {self.data_path}")
        self.spark_df = self.spark.read.csv(self.data_path, header=True, inferSchema=True)
        print(f"Loaded dataset with {self.spark_df.count()} records")
        return self.spark_df

    def get_consumption_by_appliance(self):
        """Calculate average consumption by appliance type."""
        appliances = ['Fan', 'Refrigerator', 'AirConditioner', 'Television', 'Monitor', 'MotorPump']
        result = {}

        for appliance in appliances:
            if appliance in self.spark_df.columns:
                avg = self.spark_df.select(mean(col(appliance))).collect()[0][0]
                result[appliance] = avg if avg is not None else 0
            else:
                result[appliance] = 0
                print(f"Warning: Appliance column '{appliance}' not found in dataset")

        return result

    def get_consumption_by_city(self):
        """Calculate average electricity bill by city."""
        if 'City' in self.spark_df.columns and 'ElectricityBill' in self.spark_df.columns:
            result = self.spark_df.groupBy('City').agg(mean('ElectricityBill').alias('avg_bill'))
            result = result.orderBy(desc('avg_bill'))
            return result.toPandas()
        else:
            print("Warning: 'City' or 'ElectricityBill' column not found in dataset")
            return pd.DataFrame()

    def get_consumption_by_month(self):
        """Calculate average electricity bill by month."""
        if 'Month' in self.spark_df.columns and 'ElectricityBill' in self.spark_df.columns:
            result = self.spark_df.groupBy('Month').agg(mean('ElectricityBill').alias('avg_bill'))
            return result.toPandas().sort_values('Month')
        else:
            print("Warning: 'Month' or 'ElectricityBill' column not found in dataset")
            return pd.DataFrame()

    def get_statistics(self):
        """Get basic statistics for numeric columns."""
        numeric_cols = [field.name for field in self.spark_df.schema.fields
                        if field.dataType.typeName() in ('integer', 'double', 'float', 'long')]

        stats = {}
        for col_name in numeric_cols:
            col_stats = self.spark_df.select(
                mean(col(col_name)).alias('mean'),
                min(col(col_name)).alias('min'),
                max(col(col_name)).alias('max')
            ).collect()[0]

            stats[col_name] = {
                'mean': col_stats['mean'],
                'min': col_stats['min'],
                'max': col_stats['max']
            }

        return stats

# Analysis Engine Class
class ElectricityAnalysisEngine:
    def __init__(self, spark, spark_df):
        """Initialize the analysis engine with a Spark DataFrame."""
        self.spark = spark
        self.spark_df = spark_df
        self.feature_cols = ["Fan", "Refrigerator", "AirConditioner", "Television",
                            "Monitor", "MotorPump", "MonthlyHours", "TariffRate"]
        self.available_features = [col for col in self.feature_cols if col in self.spark_df.columns]
        self.rf_model = None

        if len(self.available_features) < len(self.feature_cols):
            missing = set(self.feature_cols) - set(self.available_features)
            print(f"Warning: Missing feature columns: {missing}")

    def generate_consumption_heatmap(self):
        """Generate correlation heatmap for consumption variables."""
        all_corr_cols = ['Fan', 'Refrigerator', 'AirConditioner', 'Television',
                         'Monitor', 'MotorPump', 'MonthlyHours', 'TariffRate', 'ElectricityBill']

        corr_cols = [col for col in all_corr_cols if col in self.spark_df.columns]

        if not corr_cols:
            print("Error: No correlation columns found")
            plt.figure(figsize=(10, 8))
            plt.text(0.5, 0.5, "No correlation data available", ha='center', va='center')
            plt.tight_layout()
            return plt.gcf()

        # Convert to pandas for correlation calculation
        pdf = self.spark_df.select(corr_cols).toPandas()
        corr_matrix = pdf.corr()

        plt.figure(figsize=(10, 8))
        heatmap = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
        plt.title('Correlation Heatmap of Electricity Consumption Variables')
        plt.tight_layout()

        return heatmap.figure

    def generate_appliance_impact_chart(self):
        """Generate chart showing the impact of appliance usage on electricity bill."""
        impact_data = []

        if 'ElectricityBill' not in self.spark_df.columns:
            print("Error: ElectricityBill column not found for impact analysis")
            return []

        appliances = ['Fan', 'Refrigerator', 'AirConditioner', 'Television', 'Monitor', 'MotorPump']
        available_appliances = [app for app in appliances if app in self.spark_df.columns]

        for appliance in available_appliances:
            try:
                # Convert to pandas for analysis
                appliance_impact = self.spark_df.select(appliance, 'ElectricityBill') \
                                      .groupBy(appliance) \
                                      .agg(mean('ElectricityBill').alias('avg_bill')) \
                                      .orderBy(appliance) \
                                      .toPandas()

                impact_data.append((appliance, appliance_impact))
            except Exception as e:
                print(f"Error analyzing impact for {appliance}: {str(e)}")

        return impact_data

    def build_consumption_prediction_model(self):
        """Build predictive model for electricity consumption using RandomForest."""
        if 'ElectricityBill' not in self.spark_df.columns:
            print("Error: ElectricityBill column not found in dataset")
            return {'error': 'ElectricityBill column not found in dataset'}

        if not self.available_features:
            print("Error: No feature columns available for prediction")
            return {'error': 'No feature columns available for prediction'}

        # Prepare data for machine learning
        assembler = VectorAssembler(
            inputCols=self.available_features,
            outputCol="features"
        )

        assembled_data = assembler.transform(self.spark_df)

        # Split data into training and test sets
        train_data, test_data = assembled_data.randomSplit([0.8, 0.2], seed=42)

        # Train Random Forest model
        rf = RandomForestRegressor(
            featuresCol="features",
            labelCol="ElectricityBill",
            numTrees=100,
            maxDepth=10,
            seed=42
        )

        self.rf_model = rf.fit(train_data)

        # Make predictions
        predictions = self.rf_model.transform(test_data)

        # Evaluate model
        evaluator = RegressionEvaluator(
            labelCol="ElectricityBill",
            predictionCol="prediction",
            metricName="rmse"
        )

        rmse = evaluator.evaluate(predictions)

        evaluator.setMetricName("r2")
        r2 = evaluator.evaluate(predictions)

        evaluator.setMetricName("mae")  # Set the metric to MAE
        mae = evaluator.evaluate(predictions)

        # Get feature importance
        feature_importance = []
        for feature, importance in zip(self.available_features, self.rf_model.featureImportances):
            feature_importance.append((feature, float(importance)))

        importance_df = pd.DataFrame(feature_importance, columns=['Feature', 'Importance']) \
                          .sort_values('Importance', ascending=False)

        # Convert test data to pandas for plotting
        test_pdf = predictions.select("ElectricityBill", "prediction").toPandas()

        return {
            'model': self.rf_model,
            'rf_rmse': rmse,
            'rf_r2': r2,
            'mae_r': mae,
            'feature_importance': importance_df,
            'feature_columns': self.available_features,
            'test_data': test_pdf
        }

    def predict_electricity_charge(self): # Added this method to the class
        """Predicts the electricity charge based on user input."""
        print("\nEnter values for prediction:")
        input_data = {}
        for feature in self.available_features:
            while True:
                try:
                    value = float(input(f"Enter value for {feature}: "))
                    input_data[feature] = value
                    break
                except ValueError:
                    print("Invalid input. Please enter a numerical value.")

        input_df = pd.DataFrame([input_data])
        input_spark_df = self.spark.createDataFrame(input_df)

        # Assemble features for prediction
        assembler = VectorAssembler(inputCols=self.available_features, outputCol="features")
        assembled_input = assembler.transform(input_spark_df)

        # Make prediction
        prediction = self.rf_model.transform(assembled_input)
        predicted_charge = prediction.select("prediction").collect()[0][0]

        print(f"\nPredicted Electricity Charge: ₹{predicted_charge:.2f}")



    def identify_consumption_patterns(self, n_clusters=3):
        """Identify consumption patterns using K-means clustering."""
        # Select numerical features for clustering
        numeric_features = [col for col in self.spark_df.columns
                           if col in self.available_features]

        if not numeric_features:
            return {'error': 'No numeric features found for clustering'}

        # Assemble features into a vector
        assembler = VectorAssembler(
            inputCols=numeric_features,
            outputCol="features_unscaled"
        )

        assembled_data = assembler.transform(self.spark_df)

        # Scale the features
        scaler = StandardScaler(
            inputCol="features_unscaled",
            outputCol="features",
            withStd=True,
            withMean=True
        )

        scaler_model = scaler.fit(assembled_data)
        scaled_data = scaler_model.transform(assembled_data)

        # Apply K-means clustering
        kmeans = KMeans(
            featuresCol="features",
            predictionCol="Cluster",
            k=n_clusters,
            seed=42
        )

        kmeans_model = kmeans.fit(scaled_data)
        clustered_data = kmeans_model.transform(scaled_data)

        # Get cluster centers
        centers = kmeans_model.clusterCenters()

        # Compute statistics by cluster
        cluster_stats = clustered_data.groupBy("Cluster") \
            .agg(*[mean(col).alias(f"{col}_mean") for col in numeric_features])

        if 'ElectricityBill' in self.spark_df.columns:
            cluster_stats = clustered_data.groupBy("Cluster") \
                .agg(
                    count("*").alias("count"),
                    mean("ElectricityBill").alias("ElectricityBill_mean"),
                    min("ElectricityBill").alias("ElectricityBill_min"),
                    max("ElectricityBill").alias("ElectricityBill_max"),
                    *[mean(col).alias(f"{col}_mean") for col in numeric_features]
                )

        # Convert to pandas for visualization
        df_clustered = clustered_data.toPandas()
        cluster_stats_pdf = cluster_stats.toPandas()

        return {
            'kmeans_model': kmeans_model,
            'df_clustered': df_clustered,
            'cluster_stats': cluster_stats_pdf,
            'centers': centers
        }

# UI Functions with Enhanced Analysis

def show_dataset_overview(data_processor):
    """Show basic statistics and charts for the dataset with added textual analysis."""
    if data_processor is None or data_processor.spark_df is None:
        print("Please load data first.")
        return

    print("\n" + "="*50)
    print("DATASET OVERVIEW")
    print("="*50)

    # Basic statistics
    total_records = data_processor.spark_df.count()
    print(f"Total Records: {total_records}")

    stats = data_processor.get_statistics()
    if 'ElectricityBill' in stats:
        print(f"Average Bill: ₹{stats['ElectricityBill']['mean']:.2f}")
        print(f"Max Bill: ₹{stats['ElectricityBill']['max']:.2f}")
        print(f"Min Bill: ₹{stats['ElectricityBill']['min']:.2f}")

    # Print available columns
    print("\nAvailable Columns:")
    for col in data_processor.spark_df.columns:
        print(f"- {col}")

    # Select chart type
    print("\nSelect Chart Type:")
    print("1. Consumption by City")
    print("2. Consumption by Month")
    print("3. Appliance Impact")
    chart_choice = input("Enter chart choice (1-3): ")

    if chart_choice == "1":
        show_consumption_by_city_with_analysis(data_processor)
    elif chart_choice == "2":
        show_consumption_by_month_with_analysis(data_processor)
    elif chart_choice == "3":
        show_appliance_usage_with_analysis(data_processor)
    else:
        print("Invalid choice.")

def show_consumption_by_city_with_analysis(data_processor):
    """Show chart of consumption by city with detailed text analysis."""
    city_data = data_processor.get_consumption_by_city()

    if city_data.empty:
        print("No city data available.")
        return

    plt.figure(figsize=(12, 6))
    top_cities = city_data.head(10)

    plt.bar(top_cities['City'], top_cities['avg_bill'])
    plt.title('Average Electricity Bill by City')
    plt.ylabel('Average Bill (₹)')
    plt.xlabel('City')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Detailed text analysis
    print("\n" + "="*50)
    print("CITY CONSUMPTION ANALYSIS")
    print("="*50)

    highest_city = top_cities.iloc[0]
    lowest_city = top_cities.iloc[-1]

    # Calculate statistics for analysis
    overall_avg = city_data['avg_bill'].mean()
    city_std = city_data['avg_bill'].std()
    city_count = len(city_data)

    print(f"Analysis based on data from {city_count} cities:")
    print(f"\n1. Highest Consumption City: {highest_city['City']} (₹{highest_city['avg_bill']:.2f})")
    print(f"   - {highest_city['avg_bill'] / overall_avg:.1f}x higher than the average city")

    print(f"\n2. Lowest Consumption City: {lowest_city['City']} (₹{lowest_city['avg_bill']:.2f})")
    print(f"   - {overall_avg / lowest_city['avg_bill']:.1f}x lower than the average city")

    print(f"\n3. Overall City Statistics:")
    print(f"   - Average city consumption: ₹{overall_avg:.2f}")
    print(f"   - Standard deviation: ₹{city_std:.2f} (indicating {'high' if city_std/overall_avg > 0.3 else 'moderate' if city_std/overall_avg > 0.15 else 'low'} variability)")

    # Distribution analysis
    above_avg = len(city_data[city_data['avg_bill'] > overall_avg])
    print(f"\n4. Distribution Analysis:")
    print(f"   - {above_avg} cities ({above_avg/city_count*100:.1f}%) have above-average consumption")
    print(f"   - {city_count - above_avg} cities ({(city_count - above_avg)/city_count*100:.1f}%) have below-average consumption")

    # Insights
    print("\n5. Key Insights:")
    if city_std/overall_avg > 0.3:
        print("   - High variability in consumption across cities suggests significant differences in lifestyle, climate, or infrastructure")
    else:
        print("   - Low variability in consumption across cities suggests fewer differences in lifestyle, climate, or infrastructure")

    print(f"   - The highest consumption city uses {highest_city['avg_bill'] / lowest_city['avg_bill']:.1f}x more electricity than the lowest")
    print("   - This suggests opportunities for targeted energy efficiency programs in high-consumption cities")

    print("\n6. Recommendations:")
    print("   - Consider factors such as climate, population density, and infrastructure when comparing cities")
    print("   - High-consumption cities may benefit from energy efficiency audits and incentive programs")
    print("   - Low-consumption cities could provide best practices for energy conservation")

def show_consumption_by_month_with_analysis(data_processor):
    """Show chart of consumption by month with detailed text analysis."""
    month_data = data_processor.get_consumption_by_month()

    if month_data.empty:
        print("No monthly data available.")
        return

    plt.figure(figsize=(12, 6))
    plt.plot(month_data['Month'], month_data['avg_bill'], marker='o')
    plt.title('Average Electricity Bill by Month')
    plt.ylabel('Average Bill (₹)')
    plt.xlabel('Month')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Detailed text analysis
    print("\n" + "="*50)
    print("MONTHLY CONSUMPTION ANALYSIS")
    print("="*50)

    highest_month = month_data.loc[month_data['avg_bill'].idxmax()]
    lowest_month = month_data.loc[month_data['avg_bill'].idxmin()]

    # Calculate statistics for analysis
    annual_avg = month_data['avg_bill'].mean()
    month_std = month_data['avg_bill'].std()
    peak_to_avg_ratio = highest_month['avg_bill'] / annual_avg

    # Define seasons (adjust for your region)
    seasons = {
        'Winter': [12, 1, 2],
        'Spring': [3, 4, 5],
        'Summer': [6, 7, 8],
        'Fall': [9, 10, 11]
    }

    # Calculate seasonal averages
    seasonal_avgs = {}
    for season, months in seasons.items():
        seasonal_data = month_data[month_data['Month'].isin(months)]
        if not seasonal_data.empty:
            seasonal_avgs[season] = seasonal_data['avg_bill'].mean()

    # Find highest and lowest seasons
    if seasonal_avgs:
        highest_season = builtins.max(seasonal_avgs, key=seasonal_avgs.get)
        lowest_season = builtins.min(seasonal_avgs, key=seasonal_avgs.get)

    print(f"Analysis based on monthly consumption data:")
    print(f"\n1. Peak Consumption: Month {highest_month['Month']} (₹{highest_month['avg_bill']:.2f})")
    print(f"   - {peak_to_avg_ratio:.1f}x higher than the annual average")

    print(f"\n2. Lowest Consumption: Month {lowest_month['Month']} (₹{lowest_month['avg_bill']:.2f})")
    print(f"   - {annual_avg / lowest_month['avg_bill']:.1f}x lower than the annual average")

    print(f"\n3. Monthly Variation:")
    print(f"   - Annual average: ₹{annual_avg:.2f}")
    print(f"   - Standard deviation: ₹{month_std:.2f}")
    print(f"   - Coefficient of variation: {month_std/annual_avg:.2f} (indicating {'high' if month_std/annual_avg > 0.2 else 'moderate' if month_std/annual_avg > 0.1 else 'low'} seasonality)")

    # Seasonal analysis
    if seasonal_avgs:
        print(f"\n4. Seasonal Analysis:")
        for season, avg in seasonal_avgs.items():
            print(f"   - {season}: ₹{avg:.2f} ({avg/annual_avg*100-100:+.1f}% compared to annual average)")

        print(f"\n5. Seasonal Patterns:")
        print(f"   - Highest consumption in {highest_season} (₹{seasonal_avgs[highest_season]:.2f})")
        print(f"   - Lowest consumption in {lowest_season} (₹{seasonal_avgs[lowest_season]:.2f})")
        print(f"   - Seasonal ratio: {seasonal_avgs[highest_season]/seasonal_avgs[lowest_season]:.1f}x")

    # Identify patterns and trends
    print("\n6. Key Insights:")

    # Calculate month-to-month changes
    month_data['change'] = month_data['avg_bill'].diff()
    rising_months = month_data[month_data['change'] > 0]

    if not rising_months.empty:
        max_increase = rising_months.loc[rising_months['change'].idxmax()]
        print(f"   - Largest consumption increase occurs between month {max_increase['Month']-1} and {max_increase['Month']} (₹{max_increase['change']:.2f})")

    if peak_to_avg_ratio > 1.3:
        print("   - Significant peak-to-average ratio indicates potential for load balancing")

    if month_std/annual_avg > 0.15:
        print("   - Notable seasonal variation suggests weather-dependent consumption patterns")

    print("\n7. Recommendations:")
    print("   - Consider seasonal tariff adjustments to encourage load balancing")
    print("   - Focus energy efficiency initiatives before peak consumption months")
    print("   - Investigate opportunities for reducing seasonal consumption variations")

def show_appliance_usage_with_analysis(data_processor):
    """Show chart of appliance usage with detailed text analysis."""
    appliance_data = data_processor.get_consumption_by_appliance()

    if not appliance_data:
        print("No appliance data available.")
        return

    plt.figure(figsize=(12, 6))
    appliances = list(appliance_data.keys())
    values = list(appliance_data.values())

    plt.bar(appliances, values)
    plt.title('Average Appliance Usage')
    plt.ylabel('Average Usage Hours')
    plt.xlabel('Appliance')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Detailed text analysis
    print("\n" + "="*50)
    print("APPLIANCE USAGE ANALYSIS")
    print("="*50)

    # Sort appliances by usage
    sorted_appliances = sorted(appliance_data.items(), key=lambda x: x[1], reverse=True)

    # Calculate total usage hours
    total_hours = sum(appliance_data.values())

    print(f"Analysis based on {len(appliance_data)} appliance types:")

    print("\n1. Highest Usage Appliances:")
    for i, (appliance, hours) in enumerate(sorted_appliances[:3], 1):
        if hours > 0:
            print(f"   {i}. {appliance}: {hours:.1f} hours ({hours/total_hours*100:.1f}% of total usage)")

    print("\n2. Lowest Usage Appliances:")
    for i, (appliance, hours) in enumerate(sorted_appliances[-3:], 1):
        if hours > 0:
            print(f"   {i}. {appliance}: {hours:.1f} hours ({hours/total_hours*100:.1f}% of total usage)")

    # Energy consumption estimation (approximate)
    # Typical wattage for common appliances
    typical_wattage = {
        'AirConditioner': 1500,
        'Refrigerator': 150,
        'Fan': 75,
        'Television': 100,
        'Monitor': 70,
        'MotorPump': 750
    }

    print("\n3. Estimated Energy Consumption:")
    estimated_consumption = {}

    for appliance, hours in appliance_data.items():
        if appliance in typical_wattage and hours > 0:
            wattage = typical_wattage[appliance]
            daily_kwh = (wattage * hours) / 1000
            estimated_consumption[appliance] = daily_kwh
            print(f"   - {appliance}: ~{daily_kwh:.1f} kWh/day (based on typical {wattage}W usage)")

    # Impact analysis
    if estimated_consumption:
        total_estimated_kwh = sum(estimated_consumption.values())
        print(f"\n4. Consumption Distribution:")
        for appliance, kwh in sorted(estimated_consumption.items(), key=lambda x: x[1], reverse=True):
            print(f"   - {appliance}: {kwh/total_estimated_kwh*100:.1f}% of total estimated consumption")

    # Insights and recommendations
    print("\n5. Key Insights:")

    # Identify the dominant appliance
    if sorted_appliances and sorted_appliances[0][1] > 0:
        dominant_app = sorted_appliances[0][0]
        dominant_pct = sorted_appliances[0][1] / total_hours * 100
        if dominant_pct > 30:
            print(f"   - {dominant_app} accounts for a significant portion ({dominant_pct:.1f}%) of appliance usage")
            print(f"   - This suggests potential for energy savings by focusing on {dominant_app} efficiency")

    # Check for high AC or heating usage
    if 'AirConditioner' in appliance_data and appliance_data['AirConditioner'] > 6:
        print("   - High air conditioner usage indicates potential for significant savings through thermostat adjustment")

    print("\n6. Energy Efficiency Recommendations:")
    for appliance, hours in sorted_appliances:
        if appliance == 'AirConditioner' and hours > 6:
            print("   - Consider raising the air conditioner temperature setpoint by 1-2°C to save 5-10% on cooling costs")
        elif appliance == 'Refrigerator' and hours > 20:
            print("   - Ensure refrigerator door seals are intact and coils are clean for optimal efficiency")
        elif appliance == 'MotorPump' and hours > 2:
            print("   - Check water systems for leaks to reduce pump runtime")
        elif appliance == 'Fan' and hours > 12:
            print("   - Consider using a programmable timer for fans to reduce unnecessary operation")

def print_menu():
    """Print the main menu options."""
    print("\n" + "="*50)
    print("ELECTRICITY CONSUMPTION ANALYSIS (PySpark Version)")
    print("="*50)
    print("1. Load Data")
    print("2. Show Dataset Overview")
    print("3. Generate Correlation Heatmap")
    print("4. Analyze Appliance Impact")
    print("5. Build Prediction Model")
    print("6. Run Clustering Analysis")
    print("7. Exit")
    print("="*50)
    return input("Enter your choice (1-7): ")

def load_data_ui(spark):
    """UI for loading data."""
    print("\nLoading data...")
    print("Please upload your CSV file:")

    # Upload file
    uploaded = files.upload()

    if not uploaded:
        print("No file was uploaded.")
        return None

    file_path = list(uploaded.keys())[0]

    # Create data processor
    data_processor = ElectricityDataProcessor(spark, file_path)

    return data_processor

def generate_heatmap(analysis_engine):
    """Generate and display correlation heatmap."""
    if analysis_engine is None:
        print("Please load data first.")
        return

    print("\nGenerating correlation heatmap...")
    heatmap_fig = analysis_engine.generate_consumption_heatmap()
    plt.show()
    print("Correlation heatmap generated.")

def analyze_appliance_impact(analysis_engine):
    """Analyze and visualize the impact of appliances on the bill."""
    if analysis_engine is None:
        print("Please load data first.")
        return

    print("\nAnalyzing appliance impact...")
    impact_data = analysis_engine.generate_appliance_impact_chart()

    if not impact_data:
        print("No appliance impact data available.")
        return

    num_appliances = len(impact_data)
    rows = (num_appliances + 1) // 2
    cols = builtins.min(2, num_appliances)

    if num_appliances == 1:
        rows, cols = 1, 1

    plt.figure(figsize=(15, 10))

    for i, (appliance, data) in enumerate(impact_data):
        plt.subplot(rows, cols, i+1)
        plt.plot(data[appliance], data['avg_bill'], 'o-')
        plt.title(f'{appliance} Impact on Bill')
        plt.xlabel(f'{appliance} Usage Hours')
        plt.ylabel('Electricity Bill (₹)')
        plt.grid(True)

    plt.tight_layout()
    plt.show()
    print("Appliance impact analysis complete.")

def build_prediction_model(analysis_engine):
    """Build and evaluate prediction models."""
    if analysis_engine is None:
        print("Please load data first.")
        return

    print("\nTraining prediction model...")
    model_results = analysis_engine.build_consumption_prediction_model()

    if 'error' in model_results:
        print(f"Model training issue: {model_results['error']}")
        return



    print("\nRandom Forest Model Results:")
    print(f"RMSE: {model_results['rf_rmse']:.2f}")
    print(f"R² Score: {model_results['rf_r2']:.4f}")
    print(f"MAE (Accuracy): {model_results['mae_r']:.2f}") # Print the calculated MAE

    # Plot feature importance
    importance_df = model_results['feature_importance']
    plt.figure(figsize=(10, 6))
    importance_df = importance_df.sort_values('Importance')
    plt.barh(importance_df['Feature'], importance_df['Importance'])
    plt.title('Feature Importance for Electricity Consumption')
    plt.xlabel('Importance Score')
    plt.tight_layout()
    plt.show()

    # Plot actual vs predicted
    test_data = model_results['test_data']
    plt.figure(figsize=(10, 6))
    plt.scatter(test_data['ElectricityBill'], test_data['prediction'], alpha=0.5)
    min_val = builtins.min(test_data['ElectricityBill'].min(), test_data['prediction'].min()) # Use builtins.min
    max_val = builtins.max(test_data['ElectricityBill'].max(), test_data['prediction'].max()) # Use builtins.max
    plt.plot([min_val, max_val], [min_val, max_val], 'r--')
    plt.xlabel('Actual Bill')
    plt.ylabel('Predicted Bill')
    plt.title('Actual vs Predicted Electricity Bill')
    plt.tight_layout()
    plt.show()

    print("Random Forest prediction model trained successfully.")

    analysis_engine.predict_electricity_charge()


def run_clustering(analysis_engine):
    """Run clustering analysis to discover consumption patterns."""
    if analysis_engine is None:
        print("Please load data first.")
        return

    print("\nSelect number of clusters:")
    print("1. 2 clusters")
    print("2. 3 clusters")
    print("3. 4 clusters")
    print("4. 5 clusters")
    cluster_choice = input("Enter your choice (1-4): ")

    try:
        n_clusters = int({"1": 2, "2": 3, "3": 4, "4": 5}[cluster_choice])
    except KeyError:
        print("Invalid choice. Using 3 clusters.")
        n_clusters = 3

    print(f"\nRunning clustering with {n_clusters} clusters...")
    cluster_results = analysis_engine.identify_consumption_patterns(n_clusters=n_clusters)

    if 'error' in cluster_results:
        print(f"Clustering error: {cluster_results['error']}")
        return

    df_clustered = cluster_results['df_clustered']
    cluster_stats = cluster_results['cluster_stats']

    # Display cluster statistics
    print("\nCluster Statistics:")
    display(cluster_stats)

    # Visualization
    viz_features = []
    for col in ['ElectricityBill', 'AirConditioner', 'Fan', 'TariffRate']:
        if col in df_clustered.columns:
            viz_features.append(col)
        if len(viz_features) >= 3:
            break

    if len(viz_features) < 3:
        for col in df_clustered.columns:
            if col not in viz_features and pd.api.types.is_numeric_dtype(df_clustered[col].dtype):
                viz_features.append(col)
            if len(viz_features) >= 3:
                break

    # 3D plot using Plotly if enough features
    if len(viz_features) >= 3:
        fig = px.scatter_3d(
            df_clustered, x=viz_features[0], y=viz_features[1], z=viz_features[2],
            color='Cluster', title=f'Electricity Consumption Clusters ({n_clusters})',
            opacity=0.7
        )
        fig.update_layout(margin=dict(l=0, r=0, b=0, t=40))
        fig.show()

    # 2D plot for smaller feature sets
    elif len(viz_features) >= 2:
        fig = px.scatter(
            df_clustered, x=viz_features[0], y=viz_features[1],
            color='Cluster', title=f'Electricity Consumption Clusters ({n_clusters})',
            opacity=0.7
        )
        fig.update_layout(margin=dict(l=0, r=0, b=0, t=40))
        fig.show()
    else:
        print("Not enough features for visualization.")

    print(f"Clustering completed with {n_clusters} clusters.")

# Main function
def main():
    """Main function to run the application."""
    print("Initializing Spark...")
    spark = create_spark_session()
    print("Spark initialized.")

    data_processor = None
    analysis_engine = None

    while True:
        choice = print_menu()

        if choice == '1':
            # Load Data
            data_processor = load_data_ui(spark)
            if data_processor and data_processor.spark_df is not None:
                analysis_engine = ElectricityAnalysisEngine(spark, data_processor.spark_df)

        elif choice == '2':
            # Show Dataset Overview
            show_dataset_overview(data_processor)

        elif choice == '3':
            # Generate Correlation Heatmap
            generate_heatmap(analysis_engine)

        elif choice == '4':
            # Analyze Appliance Impact
            analyze_appliance_impact(analysis_engine)

        elif choice == '5':
            # Build Prediction Model
            build_prediction_model(analysis_engine)

        elif choice == '6':
            # Run Clustering Analysis
            run_clustering(analysis_engine)

        elif choice == '7':
            # Exit
            print("\nShutting down Spark...")
            spark.stop()
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

# Run the application
if __name__ == "__main__":
    main()
```
